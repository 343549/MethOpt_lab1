## Краткое описание алгоритма (псевдокод, без симплекс-метода)

1. Считать постановку ЗЛП из текстового файла.
2. Привести ограничения к стандартному виду:
   - `>=` умножить на −1 и получить `<=`.
   - `=` оставить как равенства.
   - Неотрицательность `x >= 0` учитывать отдельно.
3. Метод перебора вершин: выбрать набор из n активных ограничений (включая все равенства и часть неравенств/границ `x_i=0`), решить линейную систему для вершины.
4. Проверить допустимость найденной точки по всем ограничениям.
5. Значение целевой функции вычислить в допустимых вершинах и выбрать оптимальную по `max`/`min`.
6. Если допустимых вершин нет — решения нет; если вершина не найдена, но область допуска не пуста, возможна неограниченность.



## Инструкция по запуску

- Убедитесь, что находитесь в каталоге лабораторной работы.
- В файле `problem.txt` задайте постановку задачи в формате:

```
+2x1 +3x2 +1x3 +4x4 -> max
+1x1 +1x2 +1x3 +1x4 <= 10
+2x1 +1x2 -1x3 +1x4 = 8
+1x2 +2x3 +1x4 >= 5
```

- Запустите ячейки ниже: импорт, демонстрация и (опционально) пакетный прогон тестов.



````python
from simplex_method import SimplexMethod

def solve_from_file(path: str, precision: int = 8):
    sm = SimplexMethod()
    sm.load_problem(path)
    answer = sm.get_solution(precision=precision)
    return answer

print("Модуль импортирован. Функция solve_from_file готова к использованию.")
````

## Демонстрация: решение задачи из `problem.txt`

Ниже выполняется полный цикл: чтение, Фаза I (при наличии `=`), Фаза II, вывод ответа.



````python
answer = solve_from_file('problem.txt')
print('Ответ:', answer)

# Формат: [[x1, x2, ...], F]
if answer is None:
    print('Решений нет (недопустимо или неограничено).')
else:
    xs, F = answer
    print('Оптимальная точка:', xs)
    print('Значение целевой функции:', F)
````
